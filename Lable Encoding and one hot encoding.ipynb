```python
# In LableEncoding there is a one problem   ...  here it set male -- > 1   and female -- > 0  ..,and then it gives high priority to male...
# To avoid this we use OneHotEncoder  --->>  it creates different collumns according to categories...like  male, female..
#after doing LableEncoding we have to do OneHotEncoding to avoid priorites...

from sklearn.preprocessing import OneHotEncoder
OHE = OneHotEncoder(categorical_features = [1])   # it set for index no. 1 ...  categorical feeature means index...
x = OHE.fit_transform(x).toarray()
x
```
